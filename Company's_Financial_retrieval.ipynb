```python
# Python Library
import pandas as pd
import lxml as lxml
import numpy as np
import html5lib as ht
from pandas.io.html import read_html
pl1 = 'https://www.moneycontrol.com/financials/HDFCBANK/profit-lossVI/MS24/#MS24' # retrieving the data from site
pl2 = pl1.replace('#','2#') 
pl3 = pl1.replace('#','3#')
bl2 = pl2.replace('profit-lossVI','balance-sheetVI')
bl3 = bl2.replace('2#','3#')
bl1 = bl2.replace('2#','#')
p1 = read_html(pl1,  attrs={"class":"mctable1"}, index_col=0, header=0)
p1 = pd.DataFrame(p1[0])
p2 = read_html(pl2,  attrs={"class":"mctable1"}, index_col=0, header=0)
p2 = pd.DataFrame(p2[0])
p3 = read_html(pl3,  attrs={"class":"mctable1"}, index_col=0, header=0)
p3 = pd.DataFrame(p3[0])
b1 = read_html(bl1,  attrs={"class":"mctable1"}, index_col=0, header=0)
b1 = pd.DataFrame(b1[0])
b2 = read_html(bl2,  attrs={"class":"mctable1"}, index_col=0, header=0)
b2 = pd.DataFrame(b2[0])
b3 = read_html(bl3,  attrs={"class":"mctable1"}, index_col=0, header=0)
b3 = pd.DataFrame(b3[0])
index = p1.index.name
index2 = b1.index.name
pt = p1.merge(p2, on = index)
p = pt.merge(p3, on = index)
bt = b1.merge(b2, on = index2)
b = bt.merge(b3, on = index2)
p = p[:27]
p = p.drop([p1.index[0]])
b = b.drop([p1.index[0]])
fm = p.append(b)
fm = fm.sort_index(axis=1 ,ascending=True)
fm = fm.drop(fm.columns[[-3,-2,-1]], axis = 1)
fm = fm.infer_objects()
fm = fm.replace(np.nan, "")
fm
```
